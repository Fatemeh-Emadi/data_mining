```python
#first we import the neccessory modules

import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

#read data 
data=pd.read_csv("C:/Users/hp/Desktop/Exasens.csv")


data.head(10)
```

```python
#we clear the unnessory columns like ID and Diagnosis 
#clear the coulmns which they have NaN values like Unnamed:9 ,Unnamed:10 ,Unnamed:11 ,Unnamed:12 

data.drop("Unnamed: 9",axis=1,inplace=True)
data.drop("Unnamed: 10",axis=1,inplace=True)
data.drop("Unnamed: 11",axis=1,inplace=True)
data.drop("Unnamed: 12",axis=1,inplace=True)
data.drop("Diagnosis",axis=1,inplace=True)
data.drop("ID",axis=1,inplace=True)
data
```

```python
#fill missing values(NaN values) 
#everywhere in dataset is NaN it puts 1 
data.fillna(1, inplace = True)
data
```

```python

data.head(20)
```

```python
#if the age>120 or age<3 is the wrong data so should be clear it
for x in data.index: #check every index in dataset
  if data.loc[x, "Age"] > 120 or data.loc[x, "Age"] < 3:
    data.drop(x, inplace = True) #clear data if it is greater than 120 or less than 3
#clear duplicated datas 
data.drop_duplicates(inplace = True)
#if found the rows with reapiting data it clears them

data
```

```python
from sklearn.preprocessing import MinMaxScaler
#Normalization is a rescaling of the data from the original range so that all values are within the new range of 0 and 1.
# normalize dataset using the scikit-learn object MinMaxScaler.
scaler = MinMaxScaler()
# transform data
scaled = scaler.fit_transform(data)
print(scaled)
```
